# Simple Linear Regression

In this notebook linear regression will be examined and implemented from scratch. Linear regression can be used to model linear relationships between independant and dependant variables(denoted X and y). Generally the parametric equation can be written as:

\begin{equation*}
y = \theta_0 + \theta_1 X + \epsilon
\end{equation*}

The aim of this notebook is to examine methods of estimating the scalar coefficients $\theta_{0}$ and $\theta_1$ that minimises the error rate $\epsilon$. After this training has been completed the coefficients will be evaluated on new data to produce new results: 

\begin{equation*}
\hat{y} = \theta_0 + \theta_1 \hat{x}
\end{equation*}

We can start by generating data using a uniform distribution surrounding a linear equation

```python
# to make this notebook's output stable across runs
import numpy as np
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.randn(100,1)
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot(X, y, "b.")
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

Great! Now we have data X that is distributed uniformly around a straight line. The next step is estimate the coefficents of the linear regression. To do that we will need assign a metric to determine how well our coefficients fit the line. The most common way to do this is with the *least squares* criterion, or more specifically the *residual sum of squares* which takes sum of the square of the error of each instance to the fitted line: 

\begin{equation*}
RSS = (y_1-\hat{y}_1)^{2}+(y_2-\hat{y}_2)^{2}+...+(y_n-\hat{y}_n)^{2}
\end{equation*}

## Methods of estimating coefficients

### 1. The Normal Equation
Using calculus and linear algebra it can be shown that the coefficients that minimises the RSS can be expressed with the *Normal Equation*:

\begin{equation*}
\hat{\theta} = (X^{T}X)^{-1}X^{T}y
\end{equation*}

Additionally To implement the normal equation a bias vector of 1s need to be added to X. The reasoning behind this is that $\theta\cdot x=\theta_0+\theta_1 x$. The bias vector represents the 1 multipled by theta_0.

Let's implement this algorithm:

```python
X_b = np.c_[ np.ones(( 100, 1)), X] # add bias vector of 1s
theta_hat = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_hat
```

Now we have our two thetas which can be used on new data. Below we can plot the line generated from our regression:

```python
x_new = np.linspace(0,2,100) #generate values for x to plot
x_new_b = np.c_[ np.ones(( 100, 1)), x_new] #add bias vector of 1s 
y_pred = x_new_b.dot(theta_hat) #create predictions

plt.plot(x_new, y_pred, '-r')
plt.plot(X, y, "b.")
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

### 2. Moore-Penrose Pseudoinverse

Another method of estimating the scalar coefficients is with the Moore-Penrose Pseudoinverse:

\begin{equation*}
\hat{\theta} = X^+y
\end{equation*}

Where the pseudoinverse of X is defined as the matrix multiplication of three factors: $X^+ = V \Sigma^+ U^T$ which can be found using *Singular Value Decomposition (SVD)* to get $U, \Sigma, V^T$

The code is implemented beneath:

```python
U, S, V_transpose = np.linalg.svd(X_b) #SVD decomposition by numpy

U_transpose = U.T #Transpose U

S_plus = np.zeros((X_b.shape[0], X_b.shape[1])).T #Calculate pseudoinverse of sigma
S_plus[:S.shape[0], :S.shape[0]] = np.linalg.inv(np.diag(S))

V = V_transpose.T #"Untranspose" V

X_plus = V.dot(S_plus).dot(U_transpose) #Combine using formula for pseudoinverse

theta_hat_penrose = X_plus.dot(y) #predictions
theta_hat_penrose
```

The Moore Penrose is more efficient than the normal equation however it doesn't work $X^{T}X$ is not invertible which can happen if there are more features than instances
